# Sentiment Analysis on Facebook Messenger Data

```python
#File Path
path_to_file = '/Users/Krrish/Desktop/message.json'
```

```python
import json
with open(path_to_file) as file:
    chat_history = json.load(file)
```

```python
#Getting dictionary keys
chat_history.keys()
```

```python
chat_history['messages']
```

```python
import pandas as pd
```

```python
#Creating Data Frame of messages
messages = pd.DataFrame(chat_history['messages'])
```

```python
messages
```

```python
# Creating Date & Time  Column
def convert_time(timestamp):
    return pd.to_datetime(timestamp,unit='ms')
messages['date'] = messages['timestamp_ms'].apply(convert_time)
```

```python
messages
```

```python
#Creating Month & Year Column
def get_month(date):
    return date.month
def get_year(date):
    return date.year
messages['month'] = messages['date'].apply(get_month)
messages['year'] = messages['date'].apply(get_year)
```

```python
messages
```

```python
#Sentiment Analysis
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#Lexicon is a giant dcitionary of words that are scored with positive or negative this is what vader will use
nltk.download('vader_lexicon')
```

```python
sentiment_analyzer = SentimentIntensityAnalyzer()
```

```python
#Example 1
sentiment_analyzer.polarity_scores('Good Morning')
```

```python
#Example 2
sentiment_analyzer.polarity_scores('I am having a bad day')
```

```python
#Analysis on the messages
def get_polarity(text):
        return sentiment_analyzer.polarity_scores(text)['compound']
messages['sentiment'] = messages['content'].apply(get_polarity)
```

```python
messages
```

```python
#Mean values
messages.groupby('sender_name').mean()
```

```python
year_month = messages.groupby(['month', 'year', 'sender_name']).mean().reset_index()
```

```python
year_month
```

```python
#Sentiment Values of Jane for each month
jane = year_month[year_month['sender_name']== 'Jane Austen']['sentiment'].values
jane
```

```python
#Sentiment Values of Jane for each month
lewis = year_month[year_month['sender_name']== 'Lewis Carroll']['sentiment'].values
lewis
```

```python
#Visualization
import matplotlib.pyplot as plt
plt.figure(figsize = (20,10))
plt.plot(jane,label='jane')
plt.plot(lewis,label='lewis')
plt.legend
```
